
*Introdução*

Este trabalho terá como aspecto principal a análise de dados relacionados aos 17 SDGs, que são objetivos propostos pela ONU para um desenvolvimento sustentável da sociedade, para promover a paz e prosperidade. 

O objetivo de desenvolvimento sustentável escolhido é visualizar a mudança da economia nos ultimos anos e determinar se o objetivo 8, de trabalho descente e crescimento economico, está sendo atingido ou se estamos experienciando desvios hodiernalmente. A relevância desse objetivo é alta, considerando o rápido avanço da inflação, e como isso afetou os trabalhadores do mundo. Em termos de aspectos da disponibilidade do trabalho, seria importante observar quantas pessoas estão empregadas, e se estão gerando capital o suficiente para sustentar as suas famílias.

O indicador escolhido para analizar estes dados foi o de "Porcentagem de contribuidores ao mercado de trabalho (% da população total acima de 15 anos)", pois ele demontrará a taxa de emprego, a sua variação por ano e fará com que seja possível ver se a disponibilidade de empregos ao longo dos anos. Isso significa que conforme a porcentagem aumenta, melhor para a economia e a sociedade, pois significa que as pessoas coneguem contribuir para ela e serem capazes de comprar as suas necessidades (considerando um salário minimo razoável).  

Objetivo: https://sdgs.un.org/goals/goal8

Indicador: https://databank.worldbank.org/source/jobs#

```python
# Media mundial

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('default')
import seaborn as sns

df = pd.read_csv('csv/jobs.csv', sep=',', encoding='latin1', decimal='.')
df = df.drop(labels=['Series Name', 'Series Code', 'Country Code'], axis=1)
df = df.replace('(?<!\.)\.\.(?!\.)', np.NaN, regex=True)
df.head()

gdp_medio = {}
for years in range(1991,2017):
    yearStr = str(years)
    columnName = yearStr + ' [YR' + yearStr + ']'

    numbCount = 0
    calc = 0
    for numb in df[columnName]:
        try:
            calc += int(float(numb))
            numbCount += 1
        except:
            0

    if numbCount > 0:
        calc /= numbCount

    gdp_medio[yearStr] = calc

plt.style.use('default')
plt.figure()
plt.plot(gdp_medio.keys(), gdp_medio.values())
plt.xticks(range(0, len(gdp_medio), 5))
plt.yticks(range(0, 15, 2))
plt.title('Contribuidores ao mercado de trabalho por família - Media mundial')
plt.xlabel('Anos')
plt.ylabel('Porcentagem')
plt.show()
```

```python
# APS1

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('default')
import seaborn as sns

df = pd.read_csv('csv/jobs.csv', sep=',', encoding='latin1', decimal='.')
df = df.drop(labels=['Series Name', 'Series Code', 'Country Code'], axis=1)
df = df.replace('(?<!\.)\.\.(?!\.)', -1, regex=True)
df.head()

anosBrasil = df.iloc[27]
anosTop = df.iloc[59]
anosBottom = df.iloc[30]

plt.style.use('default')
plt.figure()
fig, ax = plt.subplots()

for areas in [anosBrasil, anosTop, anosBottom]:

    anos = []
    data = []
    i=0
    for ano in areas.index:
        if i >= 2:
            anos.append(int(float(ano[0:5])))
            data.append(int(float(areas[ano])))
        i+=1


    plt.plot(anos, data)

ax.legend(["Brasil", "Asia do sul", 'Bulgaria'])
plt.xticks(range(1990, 2017, 5))
plt.yticks(range(0, 100, 10))
plt.title('Contribuidores ao mercado de trabalho (Acima de 15 anos)')
plt.xlabel('Anos')
plt.ylabel('Porcentagem')
plt.show()
```
